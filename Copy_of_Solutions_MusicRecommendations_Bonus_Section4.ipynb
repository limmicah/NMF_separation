# PCA on a Spectrogram and Save the Result

```python
import numpy as np
import librosa
from sklearn.decomposition import PCA
from pathlib import Path

dynamic_dir = Path("./dataset/dynamic")
wav_files = sorted(dynamic_dir.glob("*.wav"))

all_data = []

for wav_path in wav_files:
    audio, sr = librosa.load(wav_path, sr=None)
    D = librosa.stft(audio)
    specgram_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

    # PCA over frequency bins (transpose keeps frequencies as features)
    pca_specgram = PCA(n_components=0.9).fit(specgram_db.T)
    transformed = pca_specgram.transform(specgram_db.T)

    all_data.append({
        "file": str(wav_path),
        "sample_rate": sr,
        "specgram_db": specgram_db,
        "pca_transformed": transformed,
        "pca_components": pca_specgram.components_,
        "pca_explained_variance_ratio": pca_specgram.explained_variance_ratio_,
        "pca_mean": pca_specgram.mean_,
        "pca_n_components": int(pca_specgram.n_components_),
    })

# Object array because each file can have different time-frame lengths
all_data_array = np.array(all_data, dtype=object)

print(f"Processed {len(all_data_array)} WAV files")
print(f"Example file: {all_data_array[0]['file'] if len(all_data_array) > 0 else 'None'}")
print(f"Array dtype: {all_data_array.dtype}")
```
